```python
def Area_Of_Circle(r):
    return float(math.pi * r**2)

def Area_Of_Bounding_Square(r):
    return float((2*r)**2)

def N(X, Y):
    return X/Y
```

```python
import random
import matplotlib.pyplot as plt

def monte(N, p=False):
    circle_points= 0
    square_points= 0
    x = []
    y = []
      
    # Total Random numbers generated= possible x
    # values* possible y values
    for i in range(N):
      
        # Randomly generated x and y values from a
        # uniform distribution
        # Range of x and y values is -1 to 1
        rand_x= float(np.random.uniform(-1, 1))
        rand_y= float(np.random.uniform(-1, 1))
        x.append(rand_x)
        y.append(rand_y)
      
        # Distance between (x, y) from the origin
        origin_dist= rand_x**2 + rand_y**2
      
        # Checking if (x, y) lies inside the circle
        if origin_dist<= 1:
            circle_points+= 1
      
        square_points+= 1
          
        # Estimating value of pi,
        # pi= 4*(no. of points generated inside the circle)/ (no. of points generated inside the square)
        pi = 4* circle_points/ square_points
    
    if p == True:
        fig, ax = plt.subplots()
        ax.add_artist(plt.Circle((0,0), 1, fill = False))
        plt.scatter(x,y,s=1)
    print("Final Estimation of Pi=", pi, " for N=", N)
    return pi
```

```python
answer = []
num = 0
for i in range(10):
    num += 1000
    answer.append(monte(num))
    
print(answer)
```

```python
answer = {}
num = 0
for i in range(100):
    num += 10000
    answer.update({num:monte(num)})
    
print(answer.values())
```
