```python
import numpy as np
import random
```

```python
def take_test(test):
    answers = []
    for question in test:
        answers.append(random.choice(question))
    return answers
```

```python
percents1 = []
questions = np.array(['A', 'B', 'C', 'D', 'E'])
test = []
for x in range(20):
    test.append(questions)
    
for i in range(1000000):  
    key = take_test(test)
    selected_answers = take_test(test)
    count = 0
    for x in range(20):
        if(key[x] == selected_answers[x]):
            count += 1
    percent = count/20
    percents1.append(percent)
average = sum(percents1)/1000000
average
```

```python
percents2 = []
questions = np.array(['A', 'B', 'C', 'D', 'E'])
test = []
for x in range(20):
    test.append(questions)
    
for i in range(1000000):    
    selected_answers = take_test(test)
    count = 0
    for x in range(20):
        if("E" == selected_answers[x]):
            count += 1
    percent = count/20
    percents2.append(percent)
average = sum(percents2)/1000000
average
```

```python
add = []
for k in percents2:
    if k > 0.2:
        add.append(k)
len(add)/1000000
```

```python
import matplotlib.pyplot as plt

plt.scatter(range(50),percents1)
plt.scatter(range(50),percents2, color='red')
plt.title('Grades of taking test')
plt.xlabel('attempts')
plt.ylabel('percentage correct')
plt.grid(True)
plt.show()
```
